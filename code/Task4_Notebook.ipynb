##### Isotropic gaussian distribution with standart deviation 0.5 in a 207-dimensional space on 1103 individuals

```python
chdir ("/Users/paulvercoustre/Documents/Education/Centrale_ESSEC/Geometric_Methods_in_Data_Analysis/Project/Data/Task4")

# generate 1103 individuals following an isotropic multivariate gaussian with d = 207
mean = np.zeros(207)

cov = 0.5*np.eye(207)
x = np.random.multivariate_normal(mean, cov,1103)
x = np.insert(x, [0], 207, axis = 1)
np.savetxt("gaussianS1.txt",x,delimiter=" ",fmt='%.5f')

cov = 0.5*np.eye(207)
x = np.random.multivariate_normal(mean, cov,1103)
x = np.insert(x, [0], 207, axis = 1)
np.savetxt("gaussianS2.txt",x,delimiter=" ",fmt='%.5f')
```

##### Uniform distribution of points in a 207-dimensional space on 1103 individuals

```python
x = np.random.uniform(-2.3,2.3,(1103,207))
x = np.insert(x, [0], 207, axis = 1)
np.savetxt("uniformS1.txt",x,delimiter=" ",fmt='%.5f')

x = np.random.uniform(-2.4,2.4,(1103,207))
x = np.insert(x, [0], 207, axis = 1)
np.savetxt("uniformS2.txt",x,delimiter=" ",fmt='%.5f')
```

### Distributions of distances between pairs of frame for BLN69, Gaussian distribution and Uniform distribution
##### We first run sbl-lrmsd-all-pairs.exe on the gaussian and unifrom individuals to obtain the distribution of their distances 

##### S1

```python
chdir ("/Users/paulvercoustre/Documents/Education/Centrale_ESSEC/Geometric_Methods_in_Data_Analysis/Project/Data/Task4")
S1_all_dist = np.loadtxt("S1_dist_pairwise.txt",delimiter=" ",usecols=range(3))
gaussianS1_all_dist = np.loadtxt("gaussianS1_all_distances.txt",delimiter=" ",usecols=range(3))
uniformS1_all_dist = np.loadtxt("uniformS1_all_distances.txt",delimiter=" ",usecols=range(3))
```

```python
sns.set_style("whitegrid")
pylab.xlim([-0.2,3.5])
sns.distplot(S1_all_dist[:,2], hist=False, color="red")
sns.distplot(gaussianS1_all_dist[:,2], hist=False, color="black")
sns.distplot(uniformS1_all_dist[:,2], hist=False, color="grey")
plt.yscale('log')
plt.title("Pairwise distances between pairs of frames analysis - S1")
plt.xlabel('Distances')
plt.ylabel('Frequences of distances')
plt.savefig('S1_curve_task4')
#plt.show()
```
